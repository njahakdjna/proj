```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("corona_India_latest1.csv")
df.head()
```

## PLOT BAR GRAPH OF TOTAL CONFIRMED CASES FOR DIFFRENT STATES

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x=df["Name of State / UT"],y=df["Total Confirmed cases"])
plt.xticks(rotation=90)
plt.show()
```

## PLOT BARGRAPH OF COMBINED ACTIVE AND DEATH CASES FOR DIFFRENT STATES

```python
import  matplotlib.pyplot as plt
import numpy as np
x=list(df["Name of State / UT"])
y=list(df["Active Cases"])
z=list(df["Death"])
width=0.2
p=np.arange(len(x))
p1=[i+width for i in p]
plt.figure(figsize=(12,10))
plt.xlabel("Name of State / UT",fontsize=15)
plt.ylabel("Total no of people",fontsize=15)
plt.title("combined active and death cases for different states",fontsize=15)
plt.bar(p,y,width=0.2,color="b",label="Active cases")
plt.bar(p1,z,width=0.2,color="r",label="Death cases")
plt.xticks(p+width/2,x,rotation=90)
plt.legend()
plt.show()

```
